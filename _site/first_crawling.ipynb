```python
import requests
from bs4 import BeautifulSoup as bs

page = requests.get("https://library.gabia.com/")
soup = bs(page.text, "html.parser")

elements = soup.select('div.esg-entry-content a > span')

for index, element in enumerate(elements, 1):
		print("{} 번째 게시글의 제목: {}".format(index, element.text))
```

```python
import requests
from bs4 import BeautifulSoup as bs

page = requests.get("https://sports.news.naver.com/index")
soup = bs(page.text, "html.parser")

elements = soup.select('a > div.text_area > strong')
for index, element in enumerate(elements, 1):
		print("{} 번째 게시글의 제목: {}".format(index, element.text))
```

```python
import requests
from bs4 import BeautifulSoup as bs

dic = {}

page = requests.get("https://sports.news.naver.com/index")
soup = bs(page.text, "html.parser")

elements1 = soup.select('a > div.text_area > strong')
sports = []
for element in elements1:
    sports.append(element.text)

dic['sports'] = sports
dic
```

```python

page = requests.get("https://entertain.naver.com/home")
soup = bs(page.text, "html.parser")

elements2 = soup.select('div.rank_lst a')
entertains = []
for element in elements2:
    entertains.append(element.text)

dic['entertains'] = entertains
dic
```

```python
import json

print ("dic = %s" % dic)
print ("dic type = %s" % type(dic))
print ("=======")

json_val = json.dumps(dic)

print("json_val = %s" % json_val)
print("json_val type = %s" % type(json_val))

with open('./data.json','w') as f:
    json.dump(dic,f)
```
